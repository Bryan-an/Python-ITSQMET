```python
def division(a, b):
    try:
        res = a / b
    except ZeroDivisionError:
        print("No se puede dividir para cero")
    else:
        print(res)
```

```python
division(7, 0)
```

```python
while True:
    try:
        edad = int(input("Escriba su edad:"))
        break
    except ValueError:
        print("Debes ingresar un número")
        
if edad >= 18:
    print("Eres un adulto")
else:
    print("Aún no eres un adulto")
```

```python
while True:
    try:
        cad = input("Ingrese una cadena: ")
        break
    except ValueError:
        print("Debes ingresar solo letras")
        
if cad.isalpha():
    print("Logrado")
else:
    print("Incorrecto")
```

```python
def main():
    divisor = 5
    
    for i in range(100):
        if i % divisor = 0:
            print(f'{i} es divisible para {divisor}')
            
if __name__ == '__main__':
    main()
```

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

if __name__ == '__main__':
    f = factorial(2000)
    print(f)
```

```python
import sys
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

if __name__ == '__main__':
    sys.setrecursionlimit(4000)
    f = factorial(2000)
    print(f)
```
